```python
import seaborn as sns
import matplotlib.pyplot as plt
pg=sns.load_dataset('penguins')
pg.head()
```

```python
pg["species"].unique()
```

```python
pg["island"].unique()
```

```python
pg.describe()
```

```python
pg.info()
plt.show()
```

```python
pg.nunique()
```

```python
print(pg["island"].value_counts())
```

```python
sns.barplot(x='species',y='flipper_length_mm',data=pg)
plt.show()
```

```python
sex_counts = pg['sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title("Sex Distribution")
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='body_mass_g',data=pg,hue='species')
plt.show()
```

```python
sns.pairplot(pg,hue='species')
plt.show()
```

```python
sns.pairplot(pg,hue='island')
plt.show()
```

```python
sns.heatmap(pg.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sorted_data = pg.sort_values(by='flipper_length_mm')
plt.plot(sorted_data['flipper_length_mm'].reset_index(drop=True), color='purple')
plt.title("Flipper Length (Sorted)")
plt.ylabel("Length (mm)")
plt.xlabel("Index")
plt.show()
```

```python
sns.boxplot(x='species',y='bill_depth_mm',data=pg)
plt.show()
```

```python
sns.boxplot(x='species',y='bill_length_mm',data=pg)
plt.show()
```

```python
sns.lineplot(x='species',y='flipper_length_mm',data=pg)
plt.show()
```

```python
sns.relplot(x='bill_length_mm',y='bill_depth_mm',data=pg,hue='species')
```

```python
sns.jointplot(x='bill_length_mm',y='bill_depth_mm',data=pg,hue='species')
```

```python
sns.violinplot(x='species', y='bill_depth_mm', data=pg)
plt.show()
```

```python
sns.countplot(x="island", hue="species", data=pg)
```

```python
sns.histplot(data=pg, x='bill_length_mm', kde=True)
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.xticks(rotation=45)
sns.histplot(pg['species'])
plt.show()
```
